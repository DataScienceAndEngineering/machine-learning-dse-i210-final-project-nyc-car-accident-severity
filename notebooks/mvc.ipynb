```python
def determine_class_type(row):
    if row['NUMBER OF PERSONS INJURED'] == 0 and row['NUMBER OF PERSONS KILLED'] == 0:
        return '0'
    else:
        return '1'

mvc['CLASS TYPE'] = mvc.apply(determine_class_type, axis=1)
mvc['CLASS TYPE'] = mvc['CLASS TYPE'].astype('int64')
```

```python
mvc['CLASS TYPE'].value_counts()
```

```python
mvc['CRASH DATE'] = pd.to_datetime(mvc['CRASH DATE'])
mvc['YEAR'] = mvc['CRASH DATE'].dt.year
mvc['MONTH'] = mvc['CRASH DATE'].dt.month
mvc['DAY'] = mvc['CRASH DATE'].dt.day
mvc['HOUR'] = mvc['CRASH TIME'].apply(lambda x: int(x.split(':')[0]))
mvc['MINUTE'] = mvc['CRASH TIME'].apply(lambda x: int(x.split(':')[1]))
```
